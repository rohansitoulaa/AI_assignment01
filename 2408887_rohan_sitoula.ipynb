##Task - 3 - Outlier Detection:

###1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

Define outliers using the 1.5  IQR rule.

####outlier for score

```python
score = SAdf['score']
gdp = SAdf['Log GDP per capita']
```

```python
Q1_score = np.percentile(score,25,interpolation='midpoint')
Q3_score = np.percentile(score,75,interpolation='midpoint')
IQR_score = Q3_score - Q1_score
```

```python
low_lim_score = Q1_score -1.5*IQR_score
low_lim_score
```

```python
upper_lim_score = Q3_score+1.5*IQR_score
upper_lim_score
```

```python
outlier_sadf_score = SAdf[(SAdf['score']<low_lim_score) | (SAdf['score']>upper_lim_score)][['Country name']]
outlier_sadf_score
```

####outlier for GDP:

```python
Q1_gdp = np.percentile(gdp,25,interpolation='midpoint')
Q3_gdp = np.percentile(gdp,75,interpolation='midpoint')
IQR_gdp = Q3_gdp - Q1_gdp
```

```python
low_lim_gdp = Q1_gdp -1.5*IQR_gdp
low_lim_gdp
```

```python
upper_lim_gdp = Q3_gdp+1.5*IQR_gdp
upper_lim_gdp
```

```python
sadf_outlier_gdp = SAdf[(SAdf['Log GDP per capita']<low_lim_gdp) | (SAdf['Log GDP per capita']>upper_lim_gdp)][['Country name']]
sadf_outlier_gdp
```

###2. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.

```python
color = ['red' if x in outlier_sadf_score['Country name'].values or x in  sadf_outlier_gdp['Country name'].values else 'blue' for x in SAdf['Country name']]
sns.scatterplot(data = SAdf,x='Log GDP per capita',y='score',marker='x',color=color)
plt.title('Score and Log GDP per capita', fontsize=16, fontweight='bold')
plt.xlabel('Log GDP per capita', fontsize=12)
```
